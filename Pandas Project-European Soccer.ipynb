```python
# Análisis respecto a la tabla player_stats

# No considero que valga la pena deshacernos de todos los atributos que tienen valores 
# nulos en la tabla player_stats, pues son relativamente pocos valores nulos 
# comparados con el total (836/183142).
# sería un error eliminar columnas que parecen muy interesantes para un análisis posterior
# solo por el hecho de presentar algunos valores nulos como overall_rating. 
# Claramente, las columnas con valores nulos en player_stats están relacionadas, 
#pues coinciden con el número de valores nulos que presentan (837)

# Por otro lado, sí considero que valdría la pena eliminar las columnas con más de 837 valores nulos.


player_stats_drop_cols = list(player_stats_null_cols[player_stats_null_cols > 10000].index)
player_stats = player_stats.drop(player_stats_drop_cols, axis=1)
```

```python
# Reemplazar los valores nulos que están relacionados por ceros en la tabla player_stats

player_stats[['overall_rating', 'potential', 'preferred_foot', 'defensive_work_rate', 'crossing', 'finishing', 'heading_accuracy', 'short_passing', 'dribbling', 'free_kick_accuracy', 'long_passing', 'ball_control', 'acceleration', 'sprint_speed', 'reactions', 'shot_power', 'stamina', 'strength', 'long_shots', 'aggression', 'interceptions', 'positioning', 'penalties', 'marking', 'standing_tackle', 'gk_diving', 'gk_handling', 'gk_kicking', 'gk_positioning', 'gk_reflexes']] = player_stats[['overall_rating', 'potential', 'preferred_foot', 'defensive_work_rate', 'crossing', 'finishing', 'heading_accuracy', 'short_passing', 'dribbling', 'free_kick_accuracy', 'long_passing', 'ball_control', 'acceleration', 'sprint_speed', 'reactions', 'shot_power', 'stamina', 'strength', 'long_shots', 'aggression', 'interceptions', 'positioning', 'penalties', 'marking', 'standing_tackle', 'gk_diving', 'gk_handling', 'gk_kicking', 'gk_positioning', 'gk_reflexes']].fillna(0)
```

```python
# Step 6: Remover las columnas con poca varianza en player_stats

player_stats_low_variance = []

for col in player_stats._get_numeric_data():
    minimum = min(player_stats[col])
    ninety_perc = np.percentile(player_stats[col], 90)
    if ninety_perc == minimum:
        player_stats_low_variance.append(col)
#print(player_stats_low_variance)

player_stats = player_stats.drop(player_stats_low_variance, axis=1)
```

```python
# Step 7: Manipulación de Texto tabla player_stats
player_stats.head(5)
```

```python
# Step 8: Renombrar columnas en player_stats

player_stats = player_stats.rename(columns={'overall_rating':'Calificación General', 'preferred_foot':'Pie', 'attacking_work_rate': 'Calificación en Ataque', 'defensive_work_rate': 'Calificación en Defensa'})
```

```python
cal_def_count = player_stats['Calificación en Ataque']
cal_def_count.value_counts()
```

```python
# Step 9: Filtering Records tabla player_stats

# Queremos obtener un dataframe de los jugadores con un potencial alto, cuya calificación en ataque sea de categoría alta
# que sean zurdos y cuya calificación en defensa sea mayor a 70.

player_stats_filtered = player_stats[(player_stats['potential']=='high') & 
                (player_stats['Calificación en Ataque']=='high') & 
                (player_stats['Pie']=='left') & (player_stats['Calificación en Defensa']>= 70.0)]

player_stats_filtered.head()

# Nota: podríamos ser más específicos con esta tabla filtrada y remover los outliers, pero no lo consideré
# necesario porque justamente al analizar los datos de los jugadores, quienes más nos interesan son precisamente los 
# outliers (i.e, los más destacados)
```

```python
# Step 10: Binning Numeric Variables tabla player_stats

# Podemos notar que en la tabla player_stats no hay consistencia entre la 'Calificación en Ataque'
# y la 'Calificación en Defensa', ya que la primera presenta datos de tipo categórico y la segunda de 
# tipo numérico. 

# Por ello, vamos a crear variables de tipo categórico para la 'Calificación en Defensa'. 

# El 'corte' para definir dichas categorías, lo vamos a definir utilizando un criterio de percentiles. 
# Menor a percentil 50 = low, entre p50 y p80 = medium, mayor a p80 = high. 

fifty_perc = int(np.percentile(player_stats['Calificación en Defensa'], 50))
sevfive_perc = int(np.percentile(player_stats['Calificación en Defensa'], 80))

print(fifty_perc)
print(sevfive_perc)

cutoffs = [0,74,79,100]

cal_def_labels = ['low', 'medium', 'high']
categorical_cal_def = pd.cut(player_stats['Calificación en Defensa'],cutoffs, labels=cal_def_labels)
categorical_cal_def.head(20)
```

```python
# Step 11: One-Hot Encoding Categorical Variables

# Asimismo, podemos crear un dataframe de variables dummy con el atributo 'Calificación en Ataque' para la tabla player_stats:

cal_ataque = pd.get_dummies(player_stats['Calificación en Ataque'])
cal_ataque.head(20)
```
