```python
import pandas as pd
import numpy as np
```

```python
AvocadoData = pd.read_csv('../avocado.csv')
AvocadoData.head()
```

```python
NullData = AvocadoData.isnull().sum()   #No null data
#NullData
```

```python
AvocadoData.dtypes
```

```python
#Cambiar los tipos de los valores.
AvocadoData['Date'] = pd.to_datetime(AvocadoData['Date'])
AvocadoData['Total Volume'] = AvocadoData['Total Volume'].astype(int)
AvocadoData['4046'] = AvocadoData['4046'].astype(int)
AvocadoData['4225'] = AvocadoData['4225'].astype(int)
AvocadoData['4770'] = AvocadoData['4770'].astype(int)
#AvocadoData.dtypes
AvocadoData.head()
```

```python
#AvocadoData.sort_values(by=['Date'], inplace=True, ascending=False) #LOS DATOS NO DEPENDEN DE LA FECHA, FILTRO NO RELEVANTE
#AvocadoData = AvocadoData.duplicated() #NO HAY VALORES DUPLICADOS
#AvocadoData
```

```python
#Cambiar nombre de columna para indicar el cambio en la cuantificación
AvocadoData = AvocadoData.rename(columns={"Total Bags": "Total Bags in thousands"})
AvocadoData = AvocadoData.rename(columns={"Small Bags": "Small Bags in thousands"})
AvocadoData.columns
```

```python
#Dividir totales entre millón
AvocadoData['Total Bags in thousands'] = (AvocadoData['Total Bags in thousands']/1000)
AvocadoData['Small Bags in thousands'] = (AvocadoData['Small Bags in thousands']/1000)
AvocadoData.head()
```

```python
#Eliminar la columns sin nombre  UPDATE: al cambiar los miles de total bags, la columna sin nombre desapareció
AvocadoData = AvocadoData.drop('Unnamed: 0', axis=1)
AvocadoData.columns
```

```python
#DIVIDIR TIPOS DE AGUACATE ENTRE DOS COLUMNAS, CONVENCIONAL Y ORGANIC
AvocadoData = pd.get_dummies(AvocadoData, columns=['type'], prefix=['Tipo'])
AvocadoData
```

```python
AvocadoData.Tipo_organic.value_counts()
```

```python
#Agrupar valores por año
AvocadoYears = AvocadoData.groupby('year').sum()
#Convierte Average price en promedio.
MeanPrice = AvocadoData.groupby('year').mean()
#Cambia la columna de la sumatoria por la del promedio
AvocadoYears['AveragePrice'] = MeanPrice['AveragePrice']
AvocadoYears.head()
```

```python
AvocadoData['region'].value_counts()

```

```python
#Agrupar valores por año y por region

#Agrupar valores por año
AvocadoRegion = AvocadoData.groupby(['region', 'year']).sum()
#Convierte Average price en promedio.
MeanPriceRegion = AvocadoData.groupby(['region', 'year']).mean()
#Cambia la columna de la sumatoria por la del promedio
AvocadoRegion['AveragePrice'] = MeanPriceRegion['AveragePrice']
#AvocadoRegion['year'] = AvocadoData['year']

AvocadoRegion
```

```python
AvocadoData.to_csv("AvoData.csv")
AvocadoYears.to_csv("Avocado_Years.csv")
AvocadoRegion.to_csv("Avocado_Regions.csv")
```
