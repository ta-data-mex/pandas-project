```python
#1. Através de import obtuve la librería de pandas y con pd.read_cvs se abrió y pude leer el archivo de Spotify en la computadora. 
import pandas as pd

data = pd.read_csv('Top_1_000_Songs_To_Hear_Before_You_Die.csv')
data
```

```python
#2. Hice un data frame de mi base de datos con el metodo pd.DataFrame.
data = pd.DataFrame(data)
```

```python
#3. Utilicé el método data.dtypes para visualizar el tipo de valores que contiene mi Data Frame.
data.dtypes
```

```python
#4. Utilice el método data.columns para visualizar el nombre de las columnas.
data.columns
```

```python
#5. Utilicé el método value_counts para contabilizar el número de canciones que salieron por año.  
year1 = pd.value_counts(data['YEAR'])
year1
```

```python
#6. Utilicé value_counts para visualizar el número de canciones que tiene cada artista en el Data Frame.
artist1 = pd.value_counts(data['ARTIST'])
artist1
```

```python
# 7. Utilicé value_counts para visualizar cuántas canciones existen en la columna "THEME" y así mismo pude visualizar cuántas categorías existen en esta columna. 
theme1 = pd.value_counts(data['THEME'])
theme1
```

```python
#8. NULL me ayudó para visualizar en qué columnas existían valores nulos y a cuantificarlos.
null_columns = data.isnull().sum()
null_columns[null_columns > 0]
```

```python
#9. Con el método .fillna rellene los valores vacios con un guión ("-"), pensé que era lo mejor puesto que faltaban las URL pero las canciones si están dentro de la base de datos.
data[['SPOTIFY_URL']] = data[['SPOTIFY_URL']].fillna("-")
data
```

```python
#10. Con el método column_order reordené las columnas puesto que de la forma en la que estaban originalmente no resultaba atactivo en cuanto a su visualización
column_order = ['ARTIST','TITLE','YEAR','THEME','SPOTIFY_URL']
datac = data[column_order]
datac.head(3)
```

```python
#11. El método filtered me ayudó para encontrar más facilmente a un artista, en un año y en un tema en específico de todo mi universo de datos.
filtered = datac[(data['THEME']=='Love') &  
                (data['YEAR']>= 1950) & 
                (data['ARTIST'] == 'The Beatles')]
filtered
```

```python
#12. Con el método min y max pude sacar de la base de datos el año de la canción más nueva y de la más antigua.
old_song = data[['YEAR']].min(axis=0)
new_song = data[['YEAR']].max(axis=0)
print('Canción más antigua: ', old_song)

print('Canción más nueva: ', new_song)
```

```python
#13. Con el método sort.values puede ordenar de manera ascendente el nombre de los valores de la columna "ARTISTS".
datac.sort_values(by=['ARTIST'], ascending = True)
```

```python
#14. Utilice el método de "concat" (concatenación de fila) para juntar dos categorías de datos de la misma columna, en este caso ARTIST.
The_B = datac[datac['ARTIST']=='The Beatles']
Bob_Dylan = datac[datac['ARTIST']=='Bob Dylan']

BobD_The_B = pd.concat([The_B,Bob_Dylan], axis=0)

BobD_The_B.head(20)
```

```python
#15. Utilicé el método melt para ir viendo por artista y año el tipo de categoría que tenían de la columna THEME
melted = pd.melt(datac, id_vars=['ARTIST','YEAR'], 
                 value_vars=['THEME'])
melted.head(10)
```

```python
#16. Exportar a un nuevo archivo CSV

datac.to_csv('Top_1_000_Songs_To_Hear_Before_You_DieV2.csv', index=False)
```
