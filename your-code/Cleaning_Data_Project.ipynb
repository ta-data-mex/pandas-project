```python
#4 Primero voy a tirar la columna "Unnamed:10" que no sé por qué está ahí. Ni siquiera está en el csv original
libros = libros.drop(["Unnamed: 10"], axis = 1)

#Tampoco necesitamos la variabla de isbn13 (con isbn basta)
libros = libros.drop(["isbn13"], axis = 1)
```

```python
#De aquí aprendimos que las variables "average_rating", "num_pages" son tratadas como objetos. 
#Sin embargo, quiero que "num_pages" sea un int.

#Esto quiere decir que hay registros en la columna que tienen "us-Eng" en lugar de un número. 
#Toca eliminar esos datos porque son espurios.
```

```python
#Vamos a eliminar los datos que tengan "en-US" de la columna num_pages porque son incorrectos
libros.loc[(libros['num_pages']=='en-US')] =0    #con esto reemplamos a cero los datos de num_pages que tenían "en-US"
#libros.num_pages = libros.num_pages.astype(int) #ahora nos dice que hay regitros con "en". Los tenemos que remover

libros.loc[(libros['num_pages']=='en')] =0  #con esto el describe de num_pages debería salir
libros.loc[(libros['num_pages']=='eng')]= 0

libros.describe()

#Como no sale, lo forzas a que sea interger
#libros.num_pages = libros.num_pages.astype(int)
```

```python
#Esto está interesante, pero podemos tirar también aplicando condiciones. Aqui tiramos todos los que tengan 0 páginas
libros = libros.drop(libros[libros['num_pages']==0].index)   #listo. Tambien se hubiera podido 
                                                     #libros.drop(libros[libros['num_pages']==int].index) 

libros.num_pages = libros.num_pages.astype(int)
#Recuerda guardar en la misma dataframe para que puedas recuperar cambios   
libros.describe()
```

```python
#Para que sea libro tiene que tener más de 10 páginas, borramos las que tengan menos de 10
libros = libros.drop(libros[libros['num_pages'] < 10].index)
libros.describe()
```

```python
#Ahora pasamos a la variable average_rating
#libros.describe("average_rating") 

#Describe dice que tiene valores str. Entonces hay que elminarlos.
```

```python
#Trabajamos la variable "average_ratings"

libros = libros.drop(libros[libros['average_rating']==str].index) #tira todas filas con valor de str de average_rating
libros.average_rating = libros.average_rating.astype(float) #transforma de object a float
libros.describe() 
#libros
```

```python
#Variable inglés. No nos importa la variación del inglés, por lo que vamos a homogenizarlo a inglés
#Vemos cuántas variables de inglés hay

set(libros['language_code'])
```

```python
#Hay tres variaciones de inglés. Vamos a hacerla una sola
libros['language_code'] = libros['language_code'].str.replace('-','')
set(libros['language_code'])
libros['language_code'] = libros['language_code'].str.replace('CA','')
set(libros['language_code'])
libros['language_code'] = libros['language_code'].str.replace('GB','')
libros['language_code'] = libros['language_code'].str.replace('US','')
set(libros['language_code'])


#Ok, como me equivoqué en libros['language_code'] = libros['language_code'].str.replace('en','eng'), vamos a regresar todos a en
```

```python
libros['language_code'] = libros['language_code'].str.replace('eng','en')
set(libros['language_code'])
```

```python
#La siguiente variable a limpiar es la de text_reviews. Queremos libros que han sido reseñados al menos 5 veces

libros = libros.drop(libros[libros["text_reviews_count"]<5].index)
libros.describe()
```

```python
#Verificamos los valores de las páginas 
set(libros['num_pages'])  #Todos son enteros
```

```python
import numpy as np
```

```python
#Ratings. Queremos libros que tengan un ratings_count suficientemente evelvado

#Seleccionar aqui los que estén como entre el 25% de ratigs para arriba. percentile 25%
range_ratings = max(libros["ratings_count"]) - min(libros["ratings_count"])
print(range_ratings)


rango_intercuartil_menor = np.mean(libros["ratings_count"]) - np.std(libros["ratings_count"])
print(rango_intercuartil_menor)

rango_intercuartil_mayor = np.mean(libros["ratings_count"]) + np.std(libros["ratings_count"])
print(rango_intercuartil_mayor)
```

```python
#Para quedarnos solamente con los libros que tengan una dispersión estándar abajo y arriba. 
#La media menos dispersión estándar da negativo.
#Como no quiero los libros que todo el mundo lee y sobrevalora, voy a quedarme con los que tengan hasta
#una dispersión por arriba de la media y que tengan al menos 100 rate_countings
libros = libros.drop(libros[libros["ratings_count"]<100.00].index)
libros = libros.drop(libros[libros["ratings_count"] > 148867.00].index)
libros.describe()
```

```python
#Autores de libros

set(libros["authors"])  #Con autor quiero poner " et al." después del Primer autor

#Cambiar más de un autor a autor et al.
libros['authors'] = libros['authors'].str.replace(r'[-].*', ' et al.')
set(libros["authors"])
```

```python
#contar valores de una columna
tabla = pd.DataFrame(libros.title.value_counts()).reset_index()
tabla.columns = ['nombre', 'cantidad']
tabla['unico'] = np.where(tabla.cantidad == 1, True, False)
```
