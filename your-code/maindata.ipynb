```python
#remplacamos Nan por 0 en los valores de debut y final game que no tengan debut ni final game porque no nos traen informacion nueva o relevante
allbatting[["debut", "finalGame"]] = allbatting[["debut", "finalGame"]].fillna(0)
```

```python
#hice lo mismo con las columas de valores numericos
allbatting[["RBI", "SB", "CS", "SO", "IBB", "HBP", "SH", "SF", "GIDP"]] = allbatting[["RBI", "SB", "CS", "SO", "IBB", "HBP", "SH", "SF", "GIDP"]].fillna(0)
```

```python
#volvemos a checar que columnas tienen null
leage1 = allbatting.isnull().sum()
leage1
```

```python
#primero pense en rellenar los valores faltantes en la columna de bats con una estimacion tomando en cuenta 
#el porcentaje de bateadores derechos, izquierdos y ambidiestros del total, por eso hice esos pasos 
#sacamos una lista de cuantos bateadores lo hacian con la derecha y cuantos con la izquierda y cuantos con ambas
list_bats = list(allbatting["bats"])
batsl = []
batsb = []
batsr = []
batnan = []
for x, y in enumerate(list_bats):
        if y == "R":
            batsr.append(y)
        elif y == "L":
            batsl.append(y)
        elif y == "B":
            batsb.append(y)
        else:
            batnan.append(x)
```

```python
#sacamos el total de bateadores de cada mano y comprobamos que la lista de valores nulos en la columna bats es 1926 igual que en la 
print(len(batsl))
print(len(batsr))
print(len(batsb))
print(len(batnan))
```

```python
#sacamos la proporcion
total_bats = len(batsl) + len(batsr) + len(batsb) 
total_bats
propbr = len(batsr) / total_bats
propbl = len(batsl) / total_bats
propbb = len(batsb) / total_bats
print(propbb, propbl, propbr)
#con esto veriamos cuantos de los bateadores faltantes serian zurdos, cuantos derechos y cuantos ambos
batearianl = round(len(batnan) * propbl)
batearianr = round(len(batnan) * propbr)
batearianb = round(len(batnan) * propbb)
print(batearianb, batearianl, batearianr)
#pero al final no os meti porque estaria alterando la informaciÃ³n
```
