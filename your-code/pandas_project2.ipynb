```python
# Import library
import pandas as pd
```

```python
# Read the text
package = pd.read_csv('../ndctext/package.txt',sep = '\t',encoding = 'unicode_escape')
```

```python
# See how the data is structured
package.head(2)
```

```python
# Obtain the null values that are large enough so that the columns can be assigned into a list
nullvals = package.isnull().sum()
count = package.describe().loc['count','STARTMARKETINGDATE']
unavailable_lists = list(nullvals[nullvals > count/100 ].index)
print(unavailable_lists)
# Drop that value from the list
package = package.drop(unavailable_lists[0], axis=1)
package.head()
```

```python
# Change binary values from Sample_package from N and Y to 0 and 1. 

package.loc[(package['SAMPLE_PACKAGE'] == 'N'),'SAMPLE_PACKAGE'] = 0
package.loc[(package['SAMPLE_PACKAGE'] == 'Y'),'SAMPLE_PACKAGE'] = 1
package.head()
```

```python
# Drop the NDC EXCLUDE FLAG column because it has the same value for all elements. 
package = package.drop('NDC_EXCLUDE_FLAG',axis=1)
package.head(2)
```

```python
# Modify the dates from 'Start_Marketing_Date format, from float to date'
package['STARTMARKETINGDATE'] = package['STARTMARKETINGDATE'].astype('str')
package['STARTMARKETINGDATE'] = package['STARTMARKETINGDATE'].str.replace('[.]0',' ')
package['STARTMARKETINGDATE'] = pd.to_datetime(package['STARTMARKETINGDATE'], infer_datetime_format=True)
package.head(2)
```

```python
# Change the column's names for a more professional outcome
columns = list(package.columns)
new_cols = ['Product Id', 'Product NDC', 'NDC Package Code', 'Package Description', 'Start Marketing Date', 'Sample_package']
dict_cols = {columns[i]:new_cols[i] for i in range(len(new_cols))}
package = package.rename(columns=dict_cols)
package.head(4)
```

```python
# Export to a csv file with separation as '|'
package.to_csv('Package.csv', sep = '|')
```
