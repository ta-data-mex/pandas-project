```python
import pandas as pd
```

```python
import numpy as np
```

```python
DFInicial=pd.read_csv('Books.csv')
```

```python
DFInicial.head(100)
```

```python
#Checar valores vacíos:
Columnas_vacias=DFInicial.isnull().sum()
print(Columnas_vacias)
```

```python
#Estandarizar nomenclatura del idioma
DFInicial['language_code']=DFInicial['language_code'].str.replace('en-CA', 'eng').str.replace('en-US','eng').str.replace('en-GB', 'eng')
```

```python
#Eliminar errores en los nombres de Autor(es):
#Elimar doble espacio
DFInicial['authors']=DFInicial['authors'].str.replace('  ', ' ')

#Eliminar la forma actual de enlistar dos autores: Tiago Gruenner-Paco Paquin por Tiago Gruenner, Paco Paquin
DFInicial['authors']=DFInicial['authors'].str.replace('-', ', ')
```

```python
#Revisar libros que muestren tener 0 páginas:
SinPaginas = DFInicial[(DFInicial['# num_pages']==0)]

#Convertir a lista
SinPaginas = list(SinPaginas.index)
```

```python
#Eliminar libros sin páginas
DFInicial = DFInicial.drop(SinPaginas, axis=0)
```

```python
#Eliminar libros con pocas páginas: En este caso se toma como umbral aquellos con menos de 50 páginas:

#Encontrar aquellos con pocas páginas
PocasPaginas = DFInicial[(DFInicial['# num_pages']<50)]

#Convertir a lista en función del índice
PocasPaginas = list(PocasPaginas.index)

#Eliminar libros con pocas páginas
DFInicial = DFInicial.drop(PocasPaginas, axis=0)

```

```python
#ELIMINAR LIBROS QUE CUENTEN CON UN AVERAGE RATING DIFERENTE A ZERO Y QUE A SU VEZ TENGAN UN RATING_COUNT DE 0
#Encontrar libros con un average rating mayor a 0 y a la vez un rating count de 0:
RatingImposible = DFInicial[(DFInicial['average_rating']!=0) & (DFInicial['ratings_count']==0)]

#Lista para el index preparando para el "Drop"
RatingImposible=list(RatingImposible.index)

#Eliminar las filas
DFInicial=DFInicial.drop(RatingImposible, axis=0)

print(DFInicial.loc[350:450, :])
```

```python
#Agrupar por Average Rating:
#Rating menor a 1
DF0a1 = DFInicial.loc[DFInicial['average_rating']<1]

#Rating [1,2)
DF1a2 = DFInicial[(DFInicial['average_rating']>=1) & (DFInicial['average_rating']<2)]

#Rating [2,3)
DF2a3 = DFInicial[(DFInicial['average_rating']>=2) & (DFInicial['average_rating']<3)]

#Rating [3,4)
DF3a4 = DFInicial[(DFInicial['average_rating']>=3) & (DFInicial['average_rating']<4)]

#Rating [4,5]
DF4a5 = DFInicial[(DFInicial['average_rating']>=4) & (DFInicial['average_rating']<=5)]
```

```python
DFInicial.to_csv('Books_clean.csv')
```
