## Plane Crashes Dataset 
 
Data format:  


```
Format date:    Date of accident,  in the format - January 01, 2001 
time:    Local time, in 24 hr. format unless otherwise specified
location: location information
Airline/Op:  Airline or operator of the aircraft
flight_no:   Flight number assigned by the aircraft operator
route:   Complete or partial route flown prior to the accident
ac_type:     Aircraft type
registration:    ICAO registration of the aircraft
cn_ln:   Construction or serial number / Line or fuselage number
aboard:  Total aboard (passengers / crew)
fatalities:  Total fatalities aboard (passengers / crew)
ground:  Total killed on the ground
summary:     Brief description of the accident and cause if known
```


Import useful libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re

%matplotlib inline
````

Load the data into a dataframe with pandas and show the first 10 rows to determinate the structure of the data.

````python
crashes = pd.read_csv('planecrashinfo_20181121001952.csv')
crashes.head(10) 
````

We can see several things wrong with the data:

- The data in the columns date and time has no correct format for dates
- Several columns have '?' in their registries, indicating no data
- The columns aboard and fatalities make distinction between crew and passengers, however this distinction is made in every registry, which make the data difficult to access

Next is to determine how many null values this dataframe has.

````python
crashes.isnull().sum()
````

In the cell above it is shown that the dataframe doesn't have null values, however, as previously seen in the head of the dataframe, instead it has the `?` char where there is no data.
In the Heatmap below we can see the occurrence of the `?` char in each column of the dataframe (yellow color). It is clear that there is no single registry who is completely empty.

````python
plt.figure(figsize=(20,12))
sns.set(font_scale = 2)
sns.heatmap(crashes=='?', cmap = 'viridis', cbar = False, yticklabels = False)
````

The first change to be made will be on the `date` and `time` columns; due to the nature of the data in this columns, the information can be hold in one column, wich will be the result of the merging of both columns. 

The column `date_time` is created by formating and merging the columns `date` and `time`, `date_time` has the format `YYYY-bb-dd HH:MM:SS`.

````python
crashes.date = pd.to_datetime(crashes.date)
crashes.time = pd.to_timedelta(crashes.time.str.replace(':','').apply(lambda x : re.sub(r'[^\d{4}]','',x)).apply(lambda x : x[:2]+':'+x[2:]+':00' if not x == '' else ''), errors='ignore')
crashes['date_time'] =  crashes.apply(lambda x: x.date + x.time if not pd.isnull(x.time) else x.date, axis = 1)
````

Next will be managing the data in the `aboard` and `fatalities` columns, from a simple look, it is seen that each entry has information stored in a messy format. 

In order to have a better access to data, the column `aboard` will be splitted into three columns, one for each count aboard (total, passengers and crew).

````python
crashes['total_aboard'] = crashes.aboard.str.split().str[0].apply(lambda x: int(x) if x != '?' else x)
crashes['total_passengers'] = crashes.aboard.str.split().str[1].str.split(':').str[1].apply(lambda x: int(x) if x != '?' else x)
crashes['total_crew'] = crashes.aboard.str.split().str[2].str.replace(')','').str.split(':').str[1].apply(lambda x: int(x) if x != '?' else x)
````

Given that `fatalities` has an equal format as `aboard`, the same as above will be applied to `fatalities`, again, it will be splitted into three columns (total, passengers and crew).

````python
crashes['total_fatalities'] = crashes.fatalities.str.split().str[0].apply(lambda x: int(x) if x != '?' else x)
crashes['passenger_fatalities'] = crashes.fatalities.str.split().str[1].str.split(':').str[1].apply(lambda x: int(x) if x != '?' else x)
crashes['crew_fatalities'] = crashes.fatalities.str.split().str[2].str.replace(')','').str.split(':').str[1].apply(lambda x: int(x) if x != '?' else x)
````

After all the formating and creation of new columns it's done, we replace all the `?` in the dataframe by the string `NoData` in order to advice the analyst that there was no data available for the given registry.

````python
crashes.replace({'[?]':'NoData'}, regex=True, inplace = True)
````

The former columns from which the new columns were calculated are dumped and the dataframe is rearranged.

````python
crashes.rename(columns={'ground':'ground_fatalities'}, inplace=True)
names = crashes.columns.tolist()
crashes = crashes[sum([[names[13]], names[2:9], [names[12]], names[14:], [names[11]]],[])]
crashes.head(10)
````

Finally, the dataframe is saved into a csv file called `planecrashinfo_clean.csv`

````python
crashes.to_csv('planecrashinfo_clean.csv')
````
