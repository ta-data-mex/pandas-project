### Normalizando columna 'time'

```python
#Limpiando los valores que dispersan mi muestra para determinar la hora del accidente 

crash_plane['time'] = crash_plane['time'].str.replace('?', '00:00')
crash_plane['time'] = crash_plane['time'].str.replace('c', '')
#crash_plane['time'] = crash_plane['time'].str.replace(':18', '00:00')
#crash_plane['time'] = crash_plane['time'].str.replace('17', '00:17')
print(crash_plane['time'])

#raw_data['Mycol'] =  pd.to_datetime(raw_data['Mycol'], format='%d%b%Y:%H:%M:%S.%f')


```

## Normalizando columna Date

```python

#raw_data['Mycol'] =  pd.to_datetime(raw_data['Myco, format='%d%b%Y:%H:%M:%S.%f')
crash_plane['date'] = crash_plane['date'].str.replace('January', '1')
crash_plane['date'] = crash_plane['date'].str.replace('February', '2')
crash_plane['date']= crash_plane['date'].str.replace('March', '3')
crash_plane['date'] = crash_plane['date'].str.replace('Marh', '3')
crash_plane['date'] = crash_plane['date'].str.replace('April', '4')
crash_plane['date'] = crash_plane['date'].str.replace('May', '5')
crash_plane['date'] = crash_plane['date'].str.replace('June', '6')
crash_plane['date'] = crash_plane['date'].str.replace('July', '7')
crash_plane['date'] = crash_plane['date'].str.replace('August', '8')
crash_plane['date'] = crash_plane['date'].str.replace('September', '9')
crash_plane['date'] = crash_plane['date'].str.replace('October', '10')
crash_plane['date'] = crash_plane['date'].str.replace('Otober', '10')
crash_plane['date'] = crash_plane['date'].str.replace('November', '11')
crash_plane['date'] = crash_plane['date'].str.replace('December', '12')
crash_plane['date'] = crash_plane['date'].str.replace('Deember', '12')
crash_plane['date'] = crash_plane['date'].str.replace(',','')
crash_plane['date'] = crash_plane['date'].str.replace(' ','-')
print(set(crash_plane['date']))
```

## Normalizando columna Operator

```python
#Normalizando columna Operator eliminando - y ?

crash_plane['operator'] = crash_plane['operator'].str.replace('-', '')
crash_plane['operator'] = crash_plane['operator'].str.replace('?', 'sin datos')

crash_plane['operator']
```

## Normalizando columna Flight_no

```python
crash_plane['flight_no'] = crash_plane['flight_no'].astype('str')
crash_plane['flight_no'].dtype

crash_plane['flight_no'] = crash_plane['flight_no'].replace('?','0')
crash_plane['flight_no'] = crash_plane['flight_no'].replace('F-AIKG','0')
crash_plane['flight_no'] = crash_plane['flight_no'].replace('15A','15')

crash_plane['flight_no']


#data["Age"]= data["Age"].replace(25.0, "Twenty five")
#filter = data["Age"]=="Twenty five"
  
# printing only filtered columns 
#data.where(filter).dropna()
```

## Normalizando route

```python
crash_plane['route'] = crash_plane['route'].replace('?','0')
crash_plane['route'] = crash_plane['route'].replace('0','sin datos')

crash_plane['route']
```

## Normalizando columna AcType

```python
import re

crash_plane['ac_type'] = crash_plane['ac_type'].replace('\)','')
crash_plane['ac_type'] = crash_plane['ac_type'].replace('\(','')

#crash_plane['ac_type'] = crash_plane['ac_type'].str.replace(r'\s+\(\w+\)\s+', '', regex=True)

crash_plane['ac_type']
```

## Normalizando aboard

```python
import re

crash_plane['aboard'] = crash_plane['aboard'].str.replace('(', '')
crash_plane['aboard'] = crash_plane['aboard'].str.replace(')', '')
crash_plane['aboard'] = crash_plane['aboard'].str.replace(':', '')



crash_plane['aboard'] = crash_plane['aboard'].str.replace(r'passengers\d+\s+crew\d+', '', regex=True)
crash_plane['aboard'] = crash_plane['aboard'].str.replace(r'passengers\D+\s+crew\D+', '', regex=True)
crash_plane['aboard'] = crash_plane['aboard'].str.replace(r'passengers\d', '', regex=True)
crash_plane['aboard'] = crash_plane['aboard'].str.replace(r'\s+passengers','' , regex=True)
crash_plane['aboard'] = crash_plane['aboard'].str.replace(r'\s+crew','' , regex=True)

crash_plane['aboard'] = crash_plane['aboard'].str.replace('\xa0', '')
crash_plane['aboard'] = crash_plane['aboard'].str.replace(':', '')
crash_plane['aboard'] = crash_plane['aboard'].str.replace('?', '')
crash_plane['aboard'] = crash_plane['aboard'].str.replace(' ', '')
crash_plane['aboard'] = crash_plane['aboard'].replace('',0)

#print(crash_plane['aboard'].iloc[5])
print(crash_plane['aboard'])
#crash_plane['aboard'] = crash_plane['aboard'].str.replace('passengers', '')
#crash_plane['aboard'] = crash_plane['aboard'].str.replace('crew', '')
#print(set(crash_plane['aboard']))
#crash_plane['ac_type'] = crash_plane['ac_type'].replace('\(','')
```

## Normalizando fatalidades

```python
#Me encuentro con passengers y crew que quiero eliminar y dejar solo un numero que muestre las fatalidades

crash_plane['fatalities']
```

```python
#Sigo utilizando la libreria re para dejar lista la base 

crash_plane['fatalities'] = crash_plane['fatalities'].str.replace('(', '')
crash_plane['fatalities'] = crash_plane['fatalities'].str.replace(')', '')
crash_plane['fatalities'] = crash_plane['fatalities'].str.replace(':', '')



crash_plane['fatalities'] = crash_plane['fatalities'].str.replace(r'passengers\d+\s+crew\d+', '', regex=True)
crash_plane['fatalities'] = crash_plane['fatalities'].str.replace(r'passengers\D+\s+crew\D+', '', regex=True)
crash_plane['fatalities'] = crash_plane['fatalities'].str.replace(r'passengers\d', '', regex=True)
crash_plane['fatalities'] = crash_plane['fatalities'].str.replace(r'\s+passengers','' , regex=True)
crash_plane['fatalities'] = crash_plane['fatalities'].str.replace(r'\s+crew','' , regex=True)

crash_plane['fatalities'] = crash_plane['fatalities'].str.replace('\xa0', '')
crash_plane['fatalities'] = crash_plane['fatalities'].str.replace(':', '')
crash_plane['fatalities'] = crash_plane['fatalities'].str.replace('?', '')
crash_plane['fatalities'] = crash_plane['fatalities'].str.replace(' ', '')
crash_plane['fatalities'] = crash_plane['fatalities'].replace('',0)
```
