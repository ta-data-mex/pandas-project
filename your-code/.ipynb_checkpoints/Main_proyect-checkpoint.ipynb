Después de revisar las columnas y sus tipos me doy cuenta que los nombres de las columnas con años son literalmente "integers". Así que tengo que modificarlos a caracteres para poder llamarlos en las funciones. Asimismo el tipo de las columnas de los años es objeto y los valores ausentes están representados por una cadena de ".." (dos puntos), por lo que tengo que quitar los dos puntos y reemplazarlos por algo, puede ser tanto un espacio vacío como un zero. Esto dependerá de lo que se quiera hacer con la información.

La información dentro de la base de datos "climate change" contiene muchos valores vacios y hay países en los que no se cuenta con ninguna información en ninguno de los años. Por lo que decido utilizar un subset de la base donde se encuentren sólo los paises donde hay información en todos los años distinta a zero. 

```python

cc_countries_with_data = climate_change[(climate_change['1990'] != 0) & (climate_change['1991'] != 0) & (climate_change['1992'] != 0) &
                            (climate_change['1993'] != 0) & (climate_change['1994'] != 0) & (climate_change['1995'] != 0) &
                            (climate_change['1996'] != 0) & (climate_change['1997'] != 0) & (climate_change['1998'] != 0) &
                            (climate_change['1999'] != 0) & (climate_change['2000'] != 0) & (climate_change['2001'] != 0) &
                            (climate_change['2002'] != 0) & (climate_change['2003'] != 0) & (climate_change['2004'] != 0) &
                            (climate_change['2005'] != 0) & (climate_change['2006'] != 0) & (climate_change['2007'] != 0) &
                            (climate_change['2008'] != 0) & (climate_change['2009'] != 0) & (climate_change['2010'] != 0)]

# En este codigo reinicio la indexación del subset porque originalmente contiene la anterior. 
cc_countries_with_data.reset_index(drop = True)
```

Una vez que tengo el subconjunto con todos los países que tienen información en todas las casillas entonces realizo un impresión de algunas estadísticas descriptivas y calculo el rango intercuartílico. 

```python
pd.set_option('display.float_format', '{:.7f}'.format)

descriptives = cc_countries_with_data.describe().transpose()
descriptives['IQR'] = descriptives['75%'] - descriptives['25%']
descriptives
```

En el siguiente código se calculan los valores que están afuera del rango intercuartílico y se guardan en una variable llamada "outliers". En este caso parece justificado mantener los valores extremos porque nos informarían de una condición extrema que puede afectar a otros países o acelerar el cambio climático. Para conocer a profundidad esto es necesario conocer con mayor detalle la naturaleza de los datos y a qué se refiere el indice. 

```python
outliers = pd.DataFrame(columns=cc_countries_with_data.columns)

for col in descriptives.index:
    iqr = descriptives.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = descriptives.at[col,'25%'] - cutoff
    upper = descriptives.at[col,'75%'] + cutoff
    results = cc_countries_with_data[(cc_countries_with_data[col] < lower) | 
                   (cc_countries_with_data[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
    
#outliers
```

Por último guardo en un archivo csv el subconjunto que generé, listo para hacer cálculos o sacar gráficas. 

```python
cc_countries_with_data.to_csv('countries_with_sufficient_info.csv', index = False)
```
