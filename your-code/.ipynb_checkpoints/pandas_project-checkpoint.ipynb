```python
#Import library
import pandas as pd
import re
```

```python
# Read the dataframe
product_df = pd.read_csv('../ndctext/product.txt',sep = '\t',encoding = 'unicode_escape')
```

```python
# Vizualize the data frame
product_df.head(3)
```

```python
#Obtain lenght of the dataframe
len(product_df)
```

```python
# Check for NaN values:
nan_val_columns = product_df.isnull().sum()

# Determine which columns dont work because at least 10% of the total elements are missing
useless_columns = list(nan_val_columns[nan_val_columns > len(product_df)/10].index)
useless_columns
```

```python
# Drop useless columns from dataframe
product_df = product_df.drop(useless_columns, axis=1)
product_df.head()
```

```python
# Check again for NaN values
product_df.isnull().sum()
```

```python
# Modify the resting columns of the dataframe
resting_columns = product_df.columns
actualized_resting_columns = ['Product Id','Product Ndc','Product Type','Proprietary',
                              'Non-Proprietary','Dosage Form','Route','Start Marketing Date',
                              'Marketing Category','Labeler','Substance',
                              'Active Numerator Strength','Active Ingredient Unit','Ndc Flag',
                              'Listing Record Certified Through']
# Dictionary (comprehension) to rename the values of the columns for more readable
dict_change_columns = {resting_columns[i]:actualized_resting_columns[i] for i in range(len(resting_columns))}
product_df = product_df.rename(columns=dict_change_columns)
product_df.head(2)
```

```python
# Delete troublesome fields with many Null values (more than 1000), this will make shorter the values of the database
product_df = product_df.dropna(subset=['Substance','Active Numerator Strength',
                                       'Active Ingredient Unit', 'Listing Record Certified Through', 'Start Marketing Date'
                                      ])
# Check for the Null Data
product_df.isnull().sum()
```

```python
null_displ = product_df[(product_df['Proprietary'].isnull()==True)|
                       (product_df['Non-Proprietary'].isnull()==True)|
                       (product_df['Route'].isnull()==True)]

null_displ = null_displ[['Product Id', 'Proprietary', 'Non-Proprietary', 'Route']]
```

```python
# Do research to change manually the values of the dataset that are null

# Proprietary_Name modification
product_df.loc[(product_df['Proprietary'].isnull()==True) &
               (product_df['Non-Proprietary'] =='Duloxetine delayed-release capsules'),'Proprietary'] = 'Duloxetine capsules' 
product_df.loc[(product_df['Proprietary'].isnull()==True) &
               (product_df['Non-Proprietary'] !='Duloxetine delayed-release capsules'),'Proprietary'] = 'Atropine' 

# Non_Proprietary_Name modification
product_df.loc[(product_df['Non-Proprietary'].isnull()==True) &
               (product_df['Proprietary'] =='MG DermaWash'),'Non-Proprietary'] = 'MG DermaSan' 
product_df.loc[(product_df['Non-Proprietary'].isnull()==True) &
               (product_df['Proprietary'] !='MG DermaWash'),'Non-Proprietary'] = 'Foaming Hand Sanitizer'

# Route_Name Modification
product_df.loc[(product_df['Route'].isnull()==True),'Route'] = 'ORAL'
```

```python
# Check again for Null data
product_df.isnull().sum()
```

```python
# Check Easy data to change
set(product_df['Listing Record Certified Through'])
```

```python
#Modify Manually the dates
product_df['Listing Record Certified Through'] = product_df['Listing Record Certified Through'].astype('str')
for i in range(17,21):
    product_df['Listing Record Certified Through'] = product_df['Listing Record Certified Through'].str.replace(f'20{i}1231.0',f'20{i}-12-31')
set(product_df['Listing Record Certified Through'])
```

```python
# Modify the dates from 'Start_Marketing_Date'
product_df['Start Marketing Date'] = product_df['Start Marketing Date'].astype('str')
product_df['Start Marketing Date'] = product_df['Start Marketing Date'].str.replace('[.]0',' ')
product_df['Start Marketing Date'] = pd.to_datetime(product_df['Start Marketing Date'], infer_datetime_format=True)
```

```python
# Modify the values for the Ndc_Exclude_flag to binary values
product_df.loc[(product_df['Ndc Flag'] == 'N'),'Ndc Flag'] = 0
product_df.loc[(product_df['Ndc Flag'] == 'E'),'Ndc Flag'] = 1
set(product_df['Ndc Flag'])
```

```python
# Route Modification: Read the set to see the elements in the route column, and then modify the nonsense  parameters manually
#print(set(product_df['Route']))
product_df.loc[(product_df['Route'] == 'ORAL; ORAL')|
               (product_df['Route'] == 'ORAL; ORAL; ORAL')|
               (product_df['Route'] == 'ORAL; ORAL; ORAL; ORAL')|
               (product_df['Route'] == 'ORAL; ORAL; ORAL; ORAL; ORAL')|
               (product_df['Route'] == 'ORAL; ORAL; ORAL; ORAL; ORAL; ORAL; ORAL')] = 'ORAL'

product_df.loc[(product_df['Route'] == 'SUBLINGUAL; SUBLINGUAL')|
               (product_df['Route'] == 'SUBLINGUAL; SUBLINGUAL; SUBLINGUAL')|
               (product_df['Route'] == 'SUBLINGUAL; SUBLINGUAL; SUBLINGUAL; SUBLINGUAL')] ='SUBLINGUAL'

product_df.loc[(product_df['Route'] == 'INTRAMUSCULAR; INTRAMUSCULAR; INTRAVENOUS')] = 'INTRAMUSCULAR; INTRAVENOUS'

product_df.loc[(product_df['Route'] == 'TOPICAL; TOPICAL')|
               (product_df['Route'] == 'TOPICAL; TOPICAL; TOPICAL')] = 'TOPICAL'
```

```python
# Produce a csv file with the information, the separation must
# be other than the elements used, in this case '|' was used.
product_df.to_csv('product.csv', sep = '|')
```
