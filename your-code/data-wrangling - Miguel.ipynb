```python
low_variance = []

for col in df2._get_numeric_data():
    minimum = min(df2[col])
    ninety_perc = np.percentile(df2[col], 90)
    if ninety_perc == minimum:
        low_variance.append(col)

print(low_variance)
```

```python
df2 = df2.drop(low_variance, axis=1)
```

```python
stats = df2.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outliers = pd.DataFrame(columns=df2.columns)

for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = df2[(df2[col] < lower) | 
                   (df2[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
print(outliers)
```
