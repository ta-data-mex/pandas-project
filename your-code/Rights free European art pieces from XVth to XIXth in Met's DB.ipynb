```python
import pandas as pd
import io
import requests
```

<h2> Import the DB </h2>

```python
url="https://media.githubusercontent.com/media/uaqro/openaccess/master/MetObjects.csv"
s=requests.get(url).content
messy = pd.read_csv(io.StringIO(s.decode('utf-8')))
messy.tail()
```

```python
#I want to check the pictures from the XV Century to the XIX Century from European Countries & Reigns that are public domain
#So I can download them and make a catalog with HQ images of those
#First I will try to get 'Culture' values so I stick to European Culture objects
messy
```

```python
#We'll check here for 'culture' values so I can get a look at how they're organized. We can see that there's a lot of
#inconsistencies because they're labeled with different values for probably the same value (Italian, Italian,Venice, Britain, british,london...)
#I will make a list with the probable values for the countries/reigns I'm looking dor so I can extract some data and rearrange again.
messy['Culture'].value_counts()
```

```python
srch_countries = ['Austria', 'Belgium', 'France', 'Germany', 'Greece', 'Ireland', 'Italy', 'Luxembourg', 'Netherlands', 'Portugal', 'Spain', 'Vatican City']
#All gentilices come from the roots of the countries names except for the Netherlands, whose gentilice is Dutch so I will cut the root of the words and add 'dutch'
srch_values = [i[:4] for i in srch_countries]
srch_values.append('Dutc')
print(srch_values)
culture_cats_total = messy['Culture'].unique()

```

```python
#There's floats in the categories 🤦🏼‍♂️ so we'll make all values string, erase the null values and sort those from
#the european cuntries we're looking for 🕵🏼‍♂️

culture_cats_total = [str(i) for i in culture_cats_total]
culture_cats_total = [i for i in culture_cats_total if i != 'nan']
srch_values = tuple(srch_values) #we turn the list to a tuple so 'startswith' can iterate trough it
culture_cats_total = [i for i in culture_cats_total if i.startswith(srch_values)]
```

```python
culture_cats_total
```

```python
# We have the full list of categories where european objects can be labeled, let's filter the main df:

messy_1 = messy[messy['Culture'].isin(culture_cats_total)]
print(len(messy_1))
#Still 44002 objects to evaluate (We've reduced the df to 1/10th)
messy_1['Department'].unique()
```

```python
#One good indicative is the 'Department' assigned to the piece. All objects pass trough a department and there's 
#one called 'European Sculpture and Decorative Arts' wich will probably fit well.

messy_1 = messy_1[messy_1['Department'] == 'European Sculpture and Decorative Arts']

#As we've filtered most of the database and some of the columns don't apply to what we're looking for, let's
#get rid of them:
null_columns = messy_1.isnull().sum()
null_columns
#We have a lot of columns with 100% null value, because the 'department' value is the same for all, we'll get rid of it too.

drop_cols = list(null_columns[null_columns==14441].index)
messy_1 = messy_1.drop(drop_cols, axis = 1)
messy_1 = messy_1.drop('Department', axis = 1)
messy_1
```

```python
#Let's look at the classification column to see if we can get some insights from there:
#Let's drop all Textiles, Ceramics and Metalwork
messy_1['Classification'].unique()

messy_1 = messy_1[~messy_1['Classification'].str.contains("Textiles")]
messy_1 = messy_1[~messy_1['Classification'].str.contains("Ceramic")]
messy_1 = messy_1[~messy_1['Classification'].str.contains("Metalwork")]
messy_1 = messy_1[~messy_1['Classification'].str.contains("Glass")]

messy_1
```

```python
messy_1.isnull().sum()
```

```python
#Let's check the range of years to stick to artpieces made between the XVth and XIXth Century

messy_1 = messy_1[messy_1['Object End Date'].between(1400, 1900, inclusive = True)]

```

```python
messy_1['Classification'].unique()
```

```python
paintings = ['Gesso', 'Paintings-Decorative', 'Painted Canvases', 'Drawings', 'Enamels-Painted', 'Paintings', 'Prints', 'Enamels|Miniatures']

sculpture = ['Sculpture-Bronze', 'Sculpture-Reproductions', 'Sculpture-Reproductions-Casts', 'Sculpture-Architectural', 'Woodwork', 'Sculpture-Architectural|Sculpture-Reproductions-Casts', 'Stone']

messy_sculpture = messy_1[messy_1['Classification'].isin(sculpture)]
messy_paintings = messy_1[messy_1['Classification'].isin(paintings)]

messy_sculpture = messy_sculpture.drop('Rights and Reproduction', axis = 1)
messy_paintings = messy_paintings.drop('Rights and Reproduction', axis = 1)
```

```python
sculpture = messy_sculpture[messy_sculpture['Is Public Domain'] == True]
paintings = messy_paintings[messy_paintings['Is Public Domain'] == True]
```

```python
sculpture[['Country', 'Spec Culture']] = sculpture['Culture'].str.split(',',expand=True)
sculpture
```

```python
paintings[['Country', 'Spec Culture']] = paintings['Culture'].str.split(',', expand=True)
paintings
```

```python
paintings['Country'].value_counts()
```

```python
sculpture['Country'].value_counts()
```

```python
d = {'Italian': 'Italy', 'Austrian':'Austria', 'Dutch': 'Netherlands', 'Netherlandish': 'Netherlands', 'German':'Germany'}

paintings = paintings.replace({"Country": d})
sculpture = sculpture.replace({"Country": d})
sculpture['Country'].value_counts()
```

```python
sculpture['Country'] = sculpture['Country'].replace({"Italian (Florence)": "Italy"})
```

```python
sculpture['Country'].value_counts()
```

```python
#I'm going to assign the value of 'Country' to the first country mentioned:
for k, v in d.items():
    sculpture.loc[sculpture['Country'].str.startswith(k, na=False), 'Country'] = v
```

```python
sculpture['Country'].value_counts()
```

```python
#Let's bin the dates of each piece so we have clearer information:
bins = [1400, 1500, 1600, 1700, 1800, 1900]
labels = ['XV', 'XVI', 'XVII', 'XVIII', 'XIX' ]
sculpture['Century'] = pd.cut(sculpture['Object End Date'], bins = bins, labels = labels)
paintings['Century'] = pd.cut(paintings['Object End Date'], bins = bins, labels = labels)
```

```python
sculpture.columns
```

```python
#Let's drop some columns, and rearrange them (Let's assume we only care about the item specs and not the artists)
sculpture = sculpture[['Object Number', 'Object ID', 'Object Name', 'Title', 'Century', 'Medium', 'Dimensions', 'Link Resource', 'Country']]
paintings = paintings[['Object Number', 'Object ID', 'Object Name', 'Title', 'Century', 'Medium', 'Dimensions', 'Link Resource', 'Country']]
```

```python
sculpture = sculpture.sort_values(['Country', 'Century'])
paintings = paintings.sort_values(['Country', 'Century'])
```

```python
sculpture = sculpture.reset_index()
paintings = paintings.reset_index()
```

```python
#!!!! Aaaaand we're ready to go:

sculpture.to_csv('European_Sculpture_Met.csv')
paintings.to_csv('European_Paintings_Met.csv')
```
