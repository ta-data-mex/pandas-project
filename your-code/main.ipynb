Esta variable lee la base de datos en .xls

```python
climate_change_st = pd.read_excel('climate_change_to_clean.xls')
```

Observo cuántos valores nulos tiene la base de datos.

```python
null_columns = climate_change_st.isnull().sum()
null_columns
```

A los valores nulos les agrego n/a en vez de 0, pues representan datos no existentes y el número cero confundiría el verdadero significado de la falta de datos.

```python
climate_change_nd = climate_change_st[['Country code', 'Country name','Series code', 'Series name', 'SCALE', 'Decimals', 1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011]].fillna('n/a') 
```

Reviso si aún hay valores nulos. Debido a que la cantidad era poca, no borro ninguna columna.

```python
climate_change_nd.isnull().sum()
```

Cambió los nombres de tres columnas para que sea más claro el significado de dos de ellos y para empatar el estilo de una con las demás.

```python
climate_change_nd = climate_change_nd.rename(columns={'SCALE':'Scale',
                                                      'Series code':'Subject code',
                                                      'Series name':'Subject'})
```

```python
climate_change_nd
```

En la columna Subject hay algunos errores de sintaxis que debo editar para que no se malentienda su significado.

```python
climate_change_nd['Subject'] = climate_change_nd['Subject'].str.replace('/pop./', 'population')
climate_change_nd['Subject'] = climate_change_nd['Subject'].str.replace('populationlation', 'population')
climate_change_nd['Subject'] = climate_change_nd['Subject'].str.replace('/ag.', '')
climate_change_nd['Subject'] = climate_change_nd['Subject'].str.replace(',', '')
climate_change_nd['Subject'] = climate_change_nd['Subject'].str.replace('.', '')
climate_change_nd['Subject'] = climate_change_nd['Subject'].str.replace('Aver', 'Avg.')
climate_change_nd['Subject'] = climate_change_nd['Subject'].str.replace('>1million', '> 1 million')
```

```python
climate_change_nd['Subject'] = climate_change_nd['Subject'].str.replace('125', '1.25')
```

Reviso que todos los temas en Subject estén correctamente editados. Debido a que son demasiados, resulta mejor categorizarlos según su tipo de información, para así su lectura sea más sencilla.

```python
set(climate_change_nd['Subject'])
```

Decido categorizarla por: datos de emisiones, trabajo por el cuidado del medio ambiente, servicios, salubridad, población, condiciones geoclimáticas, economía y proyectos por el ciudado del medio ambiente. 

```python
emissions = climate_change_nd[(climate_change_nd['Subject']=='CO2 emissions per capita (metric tons)') |
                                  (climate_change_nd['Subject']=='CO2 emissions per units of GDP (kg/$1000 of 2005 PPP $)') |  
                                  (climate_change_nd['Subject']=='CO2 emissions total (KtCO2)') |
                                  (climate_change_nd['Subject']=='Methane (CH4) emissions, total (KtCO2e)') |
                                  (climate_change_nd['Subject']=='Nitrous oxide (N2O) emissions total (KtCO2e)') |
                                  (climate_change_nd['Subject']=='Other GHG emissions total (KtCO2e)')]
```

```python
env_work = climate_change_nd[(climate_change_nd['Subject']=='Annex-I emissions reduction target') |
                                  (climate_change_nd['Subject']=='Annual freshwater withdrawals (% of internal resources)') |  
                                  (climate_change_nd['Subject']=='Disaster risk reduction progress score (1-5 scale; 5=best)') |
                                  (climate_change_nd['Subject']=='GHG net emissions/removals by LUCF (MtCO2e)') |
                                  (climate_change_nd['Subject']=='Hosted Clean Development Mechanism (CDM) projects') |
                                  (climate_change_nd['Subject']=='Hosted Joint Implementation (JI) projects') |
                                  (climate_change_nd['Subject']=='Issued Certified Emission Reductions (CERs) from CDM (thousands)') |
                                  (climate_change_nd['Subject']=='Latest UNFCCC national communication') |
                                  (climate_change_nd['Subject']=='NAMA submission') |
                                  (climate_change_nd['Subject']=='NAPA submission') |
                                  (climate_change_nd['Subject']=='Nationally terrestrial protected areas (% of total land area)')]
```

```python
services = climate_change_nd[(climate_change_nd['Subject']=='Access to electricity (% of total population)') |
                                  (climate_change_nd['Subject']=='Access to improved sanitation (% of total population)') |  
                                  (climate_change_nd['Subject']=='Access to improved water source (% of total population)') |
                                  (climate_change_nd['Subject']=='Paved roads (% of total roads)') |
                                  (climate_change_nd['Subject']=='Public sector mgmt & institutions avg (1-6 scale; 6=best)')]
```

```python
sanation = climate_change_nd[(climate_change_nd['Subject']== 'Child malnutrition underweight (% of under  5)') |
                                  (climate_change_nd['Subject']=='Malaria incidence rate (per 100000 people)') |  
                                  (climate_change_nd['Subject']=='Under-five mortality rate (per 1000)') |
                                  (climate_change_nd['Subject']=='Nurses and midwives (per 1000 people)')]
```

```python
population = climate_change_nd[(climate_change_nd['Subject']=='Population') |
                                  (climate_change_nd['Subject']=='Population below 5m (% of total)') |  
                                  (climate_change_nd['Subject']=='Population growth (annual %)') |
                                  (climate_change_nd['Subject']=='Population in urban lomerations > 1 million (%)') |
                                  (climate_change_nd['Subject']=='Population living below $1.25 a day (% of total)') |
                                  (climate_change_nd['Subject']=='Physicians (per 1000 people)') |  
                                  (climate_change_nd['Subject']=='Primary completion rate total (% of relevant  group)') |
                                  (climate_change_nd['Subject']=='Ratio of girls to boys in primary & secondary school (%)') |
                                  (climate_change_nd['Subject']=='Urban population') |
                                  (climate_change_nd['Subject']=='Urban population growth (annual %)')]
```

```python
geo_cli_conditions = climate_change_nd[(climate_change_nd['Subject']=='Avg annual precipitation (1961-1990 mm)') |
                                  (climate_change_nd['Subject']=='Avg daily min/max temperature (1961-1990 Celsius)') |  
                                  (climate_change_nd['Subject']=='Droughts floods extreme temps (% population avg. 1990-2009)') |
                                  (climate_change_nd['Subject']=='Land area below 5m (% of land area)')] 
```

```python
economy = climate_change_nd[(climate_change_nd['Subject']=='Ease of doing business (ranking 1-183; 1=best)') |
                                  (climate_change_nd['Subject']=='Energy use per capita (kilograms of oil equivalent)') |
                                  (climate_change_nd['Subject']=='Energy use per units of GDP (kg oil eq/$1000 of 2005 PPP $)') |
                                  (climate_change_nd['Subject']=='Foreign direct investment net inflows (% of GDP)') |  
                                  (climate_change_nd['Subject']=='GDP ($)') |
                                  (climate_change_nd['Subject']=='GNI per capita (Atlas $)') |
                                  (climate_change_nd['Subject']=='Investment in energy w/ private participation ($)') |
                                  (climate_change_nd['Subject']=='Investment in telecoms w/ private participation ($)') |
                                  (climate_change_nd['Subject']=='Investment in transport w/ private participation ($)') |
                                  (climate_change_nd['Subject']=='Investment in water/sanit w/ private participation ($)')]
```

```python
changes_projected = climate_change_nd[(climate_change_nd['Subject']=='Projected annual precipitation change (2045-2065 mm)') |
                                  (climate_change_nd['Subject']=='Projected annual temperature change (2045-2065 Celsius)') |  
                                  (climate_change_nd['Subject']=='Projected change in annual cool days/cold nights') |
                                  (climate_change_nd['Subject']=='Projected change in annual hot days/warm nights') |
                                  (climate_change_nd['Subject']=='Renewable energy target')] 
```

Esta última categoría sólo muestra si los países tienen planes a futuro por el cuidado del medio ambiente. La última columna es la que sirve, así que eliminó las demás que contienen valores nulos.

```python
changes_projected = changes_projected[['Country code', 'Country name', 'Subject code', 'Subject', 'Scale', 'Decimals', 2011]].drop_duplicates()
```

Después de que están categorizadas, las vuelvo a unir en una sola base de datos. Utilizo el outer para que los valores se acomoden en las mismas columnas, sin sobreescribir unas nuevas.

```python
climate_change = pd.merge(emissions, env_work, how='outer')
climate_change = pd.merge(climate_change, services, how='outer')
climate_change = pd.merge(climate_change, sanation, how='outer')
climate_change = pd.merge(climate_change, population, how='outer')
climate_change = pd.merge(climate_change, geo_cli_conditions, how='outer')
climate_change = pd.merge(climate_change, economy, how='outer')
climate_change = pd.merge(climate_change, changes_projected, how='outer')

```

Exporto la base de datos a un archivo .csv, separado por |, para una lectura más sencilla.

```python
climate_change.to_csv('climate_change.csv', sep='|')
```
